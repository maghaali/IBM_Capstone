### First we need import BeautifulSoup and urlopen for parse html and get url

```python
from bs4 import BeautifulSoup # This is will use for parsing HTML
from urllib.request import urlopen # This is will use to retrieve data from url


url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
page = urlopen(url) # will open url
html = page.read().decode("utf-8") # read url and decode it
soup = BeautifulSoup(html, "html.parser") # parsing contents of page as a html 
```

### In cell below we create list, read html as a text and split it per each line

```python
list1 = []
list1 = soup.get_text()
list1 = list1.split('\n')
# Code below will remove all empty items of list
while("" in list1) : 
    list1.remove("") 
```

### Now we should now index of first item that we want "M1A" and first item after our table which is "Most populated FSAs[3][edit]" so we know our index should be 1 less than it's index

```python
list1.index('Most populated FSAs[3][edit]')
```

```python
list1.index('M1A')
```

### Now we create a new list with just items that we want(all item of table without header) 

```python
list2 = list1[13:553]
```

### here I just created list of numbers from 0 with gap 3 to use it in next level

```python
import numpy as np
array=np.arange(int(0),int(len(list2)),int(3))
array
```

#### Now I used list above to create a list of 3 grouped items (each line will be line of data frame)

```python
list_final = []
for i in array:
    list_final = list_final+[list2[i:i+3]]
    
```

#### This is beautiful lest that so clear but we go further to convert it to pandas dataframe

```python
list_final
```

```python
import pandas as pd  # Import pandas to be able to create pandas dataframe
```

### Code below will convert out list to pandas dataframe

```python
list_final_pd = pd.DataFrame(list_final)
list_final_pd
```

### With code below we rename column names to appropriate one

```python
list_final_pd.rename(columns={0:'Postal Code',1:'Borough', 2:'Neighbourhood'}, inplace=True)
```

```python
list_final_pd
```

### Now because we should remove all cells with item 'Not assigned' so I just replace this value with default null value as np.NaN

```python
list_final_pd.replace('Not assigned',np.NaN, inplace=True)
list_final_pd
```

### Code below will remove all rows with null value

```python
list_final_pd = list_final_pd.dropna()
```

```python
list_final_pd
```

### Also code below will group all rows base on Postal Code column and will join all data of two similar row and seperate them with comma(,)

```python
list_final_pd_cl = list_final_pd.groupby(['Postal Code']).agg(', '.join).reset_index()
```

```python
list_final_pd_cl
```

### we can see that row with item M5A is now just one and value of Neighbourhood column joint

```python
list_final_pd_cl[list_final_pd_cl['Postal Code']=='M5A']
```

### Now we are going to save our final dataframe as a csv file

```python
list_final_pd_cl.to_csv('list_final_pd_cl.csv')
```
